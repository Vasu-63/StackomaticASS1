```python
#function for median of list
def returnmedian(numbers):
  n = len(numbers)
  sorted_numbers = sorted(numbers)
  if n % 2 == 1: #creating different case for odd and even
    return sorted_numbers[n // 2]
  else:
    mid1 = sorted_numbers[n // 2 - 1]
    mid2 = sorted_numbers[n // 2]
    return (mid1 + mid2) / 2
```

```python
#function for returning average of variable number of numbers
def average(*nums):
  if len(nums) == 0:
    return None # error handling
    return sum(nums) / len(nums)
```

```python
#function for division considering error handling
def divide(a,b):
  if b==0:
    print("Can't divide by 0")
    return None
  else: return a/b
```
